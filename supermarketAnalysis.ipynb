```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dfinir l'URL brute correcte
url = "https://raw.githubusercontent.com/MichelTCHATCHOUA/supermarket-sales-analysis/main/data/SuperMarket%20Analysis.csv"

# Charger les donnes
df = pd.read_csv(url, encoding='utf-8')  # Essaie 'utf-8' dabord. Sinon remplace par 'ISO-8859-1'

# Aperu des donnes
df.head()

# Dimensions du dataset
df.shape

# Colonnes disponibles
df.columns

# Types de donnes
df.dtypes

# Infos gnrales
df.info()

# Statistiques globales
df.describe()

sns.countplot(data=df, x="City")
plt.title("Nombre de ventes par ville")
plt.xticks(rotation=45)
plt.show()

sns.barplot(data=df, x="Product line", y="Sales", estimator='mean')
plt.title("Dpenses moyennes par catgorie")
plt.xticks(rotation=45)
plt.show()

# Convertir la colonne 'Date'
df['Date'] = pd.to_datetime(df['Date'])

# Agrger par jour
daily_sales = df.groupby('Date')['Sales'].sum()

# Tracer
daily_sales.plot(figsize=(10,5))
plt.title("Total des ventes par jour")
plt.ylabel("Montant des ventes")
plt.xlabel("Date")
plt.grid()
plt.show()
```
